`#Here I did side by side with both dataframes and hunted down everything that didn't quite jive.`

```python
hits_df2 = hits_df.replace('the', 'The', regex=True)
hits_df2 = hits_df.replace('and', 'And', regex=True)
hits_df2 = hits_df.replace('it', 'It', regex=True)
hits_df2 = hits_df.replace('of', 'Of', regex=True)
hits_df2 = hits_df.replace('Fifth', '5th', regex=True)
hits_df2 = hits_df.replace('Feak', 'Freak', regex=True)
hits_df2['artist'] = hits_df['artist'].replace(to_replace='track', value='Chubby Checker')
hits_df2['artist'] = hits_df['artist'].replace(to_replace='Steve Miller Band', value='The Steve Miller Band')
hits_df2['track'] = hits_df['track'].replace(to_replace='Woman', value='Women')
hits_df2['artist'] = hits_df['artist'].replace(to_replace='Debbie Boone', value='Debby Boone')

```
