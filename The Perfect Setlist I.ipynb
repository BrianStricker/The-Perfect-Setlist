So, the first thing we will do is get rid of the 'uri' column in the above index. It's a link that takes you directly to the Spotify application without having to go through the web page first...AND unnecessary for what we're after. Second, we will remove all rows with '0s' under 'target' as these were deemed 'unlistenable songs'(see data dictionary for explanation) so, also, not what we're after.
```python
combined_df.drop(columns=combined_df.columns[2], inplace=True)
combined_df
combined_df.head()
```
